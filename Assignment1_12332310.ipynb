## PS1_2 Matrix multiplication

```python
import numpy as np 
M=np.random.randint(0,51,50) 
M1=M.reshape(5,10) 
N=np.random.randint(0,51,50) 
M2=N.reshape(10,5) 
print(M1,M2) 
def Matrix_multip(M1,M2): 
    a,b=M1.shape 
    b,c=M2.shape 
    C=np.zeros((a,c)) 
    for i in range(a): 
        for j in range(c): 
            for k in range(b): 
                C[i,j]+=M1[i,k]*M2[k,j] 
    return C 
Matrix_multip(M1, M2) 
print(Matrix_multip(M1, M2))
```

## PS1_3 Pascal triangle

```python
def Pascal_triangle(K): 
    l=[]  
    for i in range(K): 
        if i==0: 
            l.append([1])   
        elif i==1: 
            l.append([1,1]) 
        else: 
            y = []   
            for j  in range(i+1): 
                if j==0 or j==i: 
                    y.append(1) 
                else: 
                    y.append(l[i-1][j]+l[i-1][j-1]) 
            l.append(y) 
    return l 
K=12  
x=Pascal_triangle(K) 
for i in range(len(x)):#逐行打印结果 
    print(x[i]) 
```

## PS1_4 Add or double

```python
def Least_moves(x): 
    current_amount = 1  
    steps = 0  # 步数 
 
    while current_amount < x: 
        if current_amount * 2 <= x: 
            current_amount *= 2  
        else: 
            current_amount += 1  
        steps += 1  # 步数加一 
 
    return steps 
 
Least_moves(2)
Least_moves(2)
```

## PS1_5 Dynamic programming

```python
import numpy as np 
def Find_expression(target, expression="", current_num=0, next_num=1): 
    if next_num == 10:   
        if current_num == target: 
#             print(expression) 
            print(expression,'=',eval(expression)) #助教测试
        return
    Find_expression(target, expression + "+" + str(next_num), current_num + next_num, next_num + 1) 
    Find_expression(target, expression + "-" + str(next_num), current_num - next_num, next_num + 1) 
    Find_expression(target, expression + str(next_num), int(str(current_num) + str(next_num)), next_num + 1) 

for i in range(1, 101): 
    print("Target:", i) 
    Find_expression(i) 
    print() 

from collections import Counter 
Total_solutions = [] 
for i in range(1, 101): 
    c = Counter(Find_expression(i),1) 
    Total_solutions.append(c) 
    print() 

print("Total_solutions:", Total_solutions) 
 
max_count = max(Total_solutions) 
min_count = min(Total_solutions) 
max_index = Total_solutions.index(max_count) + 1 
min_index = Total_solutions.index(min_count) + 1 
 
print("The maximum number of solutions is {max_count}, which occurs for the number {max_index}.") 
print("The minimum number of solutions is {min_count}, which occurs for the number {min_index}.")
```

```python
def Find_expression(n): 
    str = ["1", "2", "3", "4", "5", "6", "7", "8", "9"] 
    sym = ["+", "-", ""] 
    sum_n = 0
    for a in sym: 
        for b in sym: 
            for c in sym: 
                for d in sym: 
                    for e in sym: 
                        for f in sym: 
                            for g in sym: 
                                for h in sym: 
                                    sum = str[0] + a + str[1] + b + str[2] + c + str[3] + d + str[4] + e + str[5] + f + \
                                    str[6] + g + str[7] + h + str[8] 
                                    if eval(sum) == n: 
                                        sum_n = sum_n + 1
                                        print(sum + "=", end = '') 
                                        print(n) 
    return sum_n
```

```python
def find_num(item, count_list): # 用于找出目标数字的在list中的位置
    idx_list = []
    for index, element in enumerate(count_list):
        if element == item:
            idx_list.append(index+1)
#             print(index+1, end=" ")
    return idx_list

# 找到表达式最少的数字
num_minexpr = find_num(min(Total_solutions),Total_solutions)
print("num with Min exprs: "+str(num_minexpr)+", the amount of exprs: "+str(min(Total_solutions)))

# 找到表达式最多的数字
num_maxexpr = find_num(max(Total_solutions),Total_solutions)
print("num with Max exprs: "+str(num_maxexpr)+", the amount of exprs: "+str(max(Total_solutions)))
```
